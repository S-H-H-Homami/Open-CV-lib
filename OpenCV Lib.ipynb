Rotation

```python
row ,col , ch=img5.shape
M=cv2.getRotationMatrix2D((col/2,col/2),90,3) #(Center of rotation,rotation degree, zoom)
rotated=cv2.warpAffine(img5,M,(row,col))
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
```

```python
points1=np.float32([[20,10],[200,150],[200,200]])
points2=np.float32([[10,20],[150,200],[250,250]])
M2=cv2.getAffineTransform(points1,points2)
rotated2=cv2.warpAffine(img5,M2,(row,col))
plt.imshow(cv2.cvtColor(rotated2, cv2.COLOR_BGR2RGB))
```

Perspective

```python
points3=np.float32([[0,0],[10,0],[200,250],[300,150]])
points4=np.float32([[20,10],[30,20],[250,300],[200,50]])
M3=cv2.getPerspectiveTransform(points3,points4)
prspctv=cv2.warpPerspective(img5,M3,(row,col))
plt.imshow(cv2.cvtColor(prspctv, cv2.COLOR_BGR2RGB))
```

Template Matching

```python
img11=cv2.imread("/content/DSC00125.jpg")
img12=img11.copy()
template=cv2.imread("/content/template.jpg")

col,row,ch=template.shape

methods=[cv2.TM_CCOEFF,cv2.TM_CCOEFF_NORMED,cv2.TM_CCORR,cv2.TM_CCORR_NORMED,cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]
count=1

for method in methods:
  img11=img12.copy()
  res=cv2.matchTemplate(img11,template,method)

  min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

  if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
    top_left=min_loc
  else:
    top_left=max_loc
  bottom_right=(top_left[0]+col,top_left[1]+row)

  cv2.rectangle(img11,top_left,bottom_right,255,10)

  plt.subplot(6,2,count)
  plt.title("Javad")
  plt.imshow(cv2.cvtColor(template, cv2.COLOR_BGR2RGB))

  plt.subplot(6,2,count+1)
  plt.title("Detect")
  plt.imshow(cv2.cvtColor(img11, cv2.COLOR_BGR2RGB))

  count+=2
    
```

```python
res=cv2.matchTemplate(img11,template,cv2.TM_CCOEFF_NORMED)

threshold=0.5
loc=np.where(res>=threshold)

for pt in zip(*loc[::-1]):
  cv2.rectangle(img12,pt,(pt[0]+col,pt[1]+row),255,5)

plt.imshow(cv2.cvtColor(img12, cv2.COLOR_BGR2RGB))
```
